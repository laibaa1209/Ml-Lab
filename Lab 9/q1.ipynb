# Binary logistic regression
Binary logistic regression is used to predict the probability of a binary outcome, such as yes or no,
true or false, or 0 or 1. For example, it could be used to predict whether a customer will churn or not,
whether a patient has a disease or not, or whether a loan will be repaid or not.

# Multinomial logistic regression
Multinomial logistic regression is used to predict the probability of one of three or more possible
outcomes, such as the type of product a customer will buy, the rating a customer will give a product,
or the political party a person will vote for.

# Ordinal logistic regression
Ordinal Logistic Regression is used to predict the probability of an outcome that falls into a
predetermined order, such as the level of customer satisfaction, the severity of a disease, or the stage
of cancer.

In scikit-learn's Logistic Regression, the multi_class='auto' parameter is used as default to
automatically determine the multi-class strategy based on the nature of the problem and the solver
selected.

The solver parameter also plays a role in determining the behavior. If solver='liblinear', it is designed for binary classification and uses the 'ovr' strategy for multiclass problems. If solver is set to other
algorithms like 'lbfgs', 'newton-cg', or 'sag', logistic regression
The choice of the solver depends on various factors, including the size of the dataset, the number of
features, and the regularization requirements. Here are some considerations:

• For small to medium-sized datasets, 'liblinear' is often a good choice.

• For high-dimensional datasets, 'lbfgs' or 'newton-cg' might be more efficient.

• For large datasets, especially when the number of samples is much larger than the number of
features, 'sag' or 'saga' could be beneficial.

```python
x = df.drop('target', axis=1)
y = df['target']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
modelLR = LogisticRegression()
modelLR.fit(x_train, y_train)

y_pred = modelLR.predict(x_test)
```

```python
print("The accuarcy score is:", accuracy_score(y_test, y_pred))

print(modelLR.coef_)
print(modelLR.intercept_)
```

```python
# multinomial logistic regression can be used when the target variable has more than two classes.
lr = LogisticRegression(multi_class='multinomial', solver="newton-cg")
modelLR.fit(x_train, y_train)
```

```python
y_pred = modelLR.predict(x_test)
print("The accuarcy score for multinomial logistic regression is:", accuracy_score(y_test, y_pred))
```

1. Solvers in Logistic Regression

Scikit-learn provides several solvers for fitting Logistic Regression models. Each uses a different optimization algorithm to find the best coefficients. Common solvers:

liblinear

Based on coordinate descent (good for small datasets).

Works well with binary classification.

Supports L1 and L2 regularization.

Pros: Fast for small datasets, simple.

Cons: Slower for large datasets, doesn’t scale well with many features.

lbfgs

Uses quasi-Newton optimization (approximates Hessian).

Supports multiclass classification natively (softmax).

Pros: Good default solver, efficient for medium-sized datasets.

Cons: Can be slower than saga for very large datasets.

newton-cg

Uses Newton’s method with Conjugate Gradient.

Supports multiclass classification.

Pros: Accurate, handles L2 regularization.

Cons: Memory-intensive for very large datasets.

sag (Stochastic Average Gradient)

A stochastic gradient descent variant.

Efficient for very large datasets.

Pros: Fast for large n_samples.

Cons: Needs feature scaling, not ideal for small datasets.

saga

Variant of sag that supports L1 and elastic-net penalties.

Efficient for large datasets and sparse data.

Pros: Handles large datasets and multiclass; supports L1 regularization.

Cons: Needs more iterations for small datasets; slower than liblinear for small datasets.

newton-cholesky (less common, specialized solver)

Uses Newton’s method with Cholesky decomposition.

Rarely used; suitable for numerically stable computations.

Pros: Can be precise; mostly academic interest.

Cons: Not optimized for very large dataset
