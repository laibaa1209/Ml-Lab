```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# classification

```python
df = pd.read_csv('heart.csv')
print(df.head())

x = df.drop('target', axis=1)
y = df['target']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
modelMLP = Sequential()
modelMLP.add(Dense(units=20, input_dim=13, activation="relu"))
modelMLP.add(Dense(units=10, activation="relu"))
modelMLP.add(Dense(units=1, activation="sigmoid"))
modelMLP.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
modelMLP.fit(x_train, y_train, epochs=10, batch_size=10, validation_data=(x_test, y_test))
```

```python
y_pred = modelMLP.predict(x_test)
y_pred_binary = np.round(y_pred)
```

```python
acc_score = accuracy_score(y_test, y_pred_binary)
print("The accuracy score is:", acc_score)
```

## Regression

```python
df = pd.read_csv("Housing.csv")
df.head()
x = df.drop(columns=['price'])
y = df['price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train = x_train.apply(pd.to_numeric, errors='coerce')
x_test = x_test.apply(pd.to_numeric, errors='coerce')
y_train = pd.to_numeric(y_train, errors='coerce')
y_test = pd.to_numeric(y_test, errors='coerce')
```

```python
MLP = Sequential()
MLP.add(Dense(64, input_dim=x.shape[1], activation='relu'))
MLP.add(Dense(34, activation="relu"))
MLP.add(Dense(1, activation='linear'))
MLP.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
MLP.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))
```

```python
print(x_train.dtypes)
print(y_train.dtypes)
```

```python
y_pred = MLP.predict(x_test)
y_pred_binary = np.round(y_pred)
```

```python
print(f"accuray: ", accuracy_score(y_test, y_pred_binary))
```
