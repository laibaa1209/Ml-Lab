```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
train_data = pd.read_csv("occupancy_train.csv")
test_data = pd.read_csv("occupancy_test.csv")

train_data = train_data.drop(columns=["S.No", 'date', 'Temperature', 'CO2'])
test_data = test_data.drop(columns=["S.NO", 'date', 'Temperature', 'CO2'])
```

```python
train_data.head()
```

```python
test_data
```

```python
x_train = train_data.drop(columns=['Occupancy'])
y_train = train_data['Occupancy']

x_test = test_data.drop(columns=['Occupancy'])
y_test = test_data['Occupancy']
```

```python
results = {}
for i in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    results[i] = accuracy

print(results)

#for maximum accuraccy
best_ac = max(results.values())
best_k = max(results, key=results.get)
print(f"Best accuracy is {best_ac} for k={best_k}")
```
