```python
# b.
df = pd.read_csv("cancer.csv")

# c.
y = df['Level']
unqiue_categories = y.unique()
print("Unique categories in 'Label':", unqiue_categories)
```

```python
# d.
print("The total number of columns in the dataset:", df.shape[1])
print("first 50 rows of the dataset:\n", df.head(50))
print("bottom 30 records of te dataset:", df.tail(30))
print("The information related to memory size is: ", df.info(memory_usage="deep"))
print("The shape of the dataset is: ", df.shape)
print("no of rows", df.shape[0]) #rows
print("no of columns", df.shape[1]) #columns
print("the size of data set is: ", df.size)
```

```python
# e.
print("Number  of columns and samples in the dataset:", df.shape[1], df.shape[0])
```

```python
# f.
num_of_records = df['Level'].value_counts()
print("Numbr of records in each category of Level is: ", num_of_records)
```

```python
plt.style.use("dark_background")
plt.grid(axis='y', linestyle="--", linewidth=0.5, color="#5C544A", alpha=0.4)

colour_palette = ["#4B3621", "#5A4E3C", "#2E2A24", "#6E5849", "#3C2F2F"]

sns.countplot(
    x='Level',
    data=df,
    palette=colour_palette,
    width=0.6
)

plt.title("Records per Target Category", fontsize=16, fontweight='bold', color="#E6D3B3", pad=20)
plt.xlabel("Target Categories", fontsize=12, color="#CBBFA2", labelpad=10)
plt.ylabel("Count", fontsize=12, color="#CBBFA2", labelpad=10)

plt.show()
```

```python
# g. 
sample_records = df.iloc[550:900, df.columns.get_loc('Gender') : df.columns.get_loc('Coughing of Blood') + 1]
print("Sample records (rows 550 to 900 for specific columns):", sample_records)
```

```python
sample_records.to_csv('sampled_records.csv', index=False)
# statistical analysis
print('The statistical analysis of extracted sample is:', sample_records.describe())
print('Mode and Median:', sample_records.median(), sample_records.mean())
```

```python
# h.
sample_data = df.iloc[-20:, df.columns.get_loc('Shortness of Breath') : df.columns.get_loc('Frequent Cold') + 1]
print("Sample data (last 20 rows for specific columns:))", sample_data)

sample_data.to_csv('last_20_records.csv', index=False)
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(20,16))
sns.heatmap(corr, annot=True, cmap="YlOrBr",linewidths=0.5, square=True)
plt.title("Correlation Heatmap of Selected Records")
plt.show()
```

```python
# Distribution of Age
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(df["Age"], kde=True, bins=10, color="brown")
plt.title("Distribution of Age")

# Distribution of Frequent Cold
plt.subplot(1,2,2)
sns.histplot(df["Frequent Cold"], kde=True, bins=10, color="beige")
plt.title("Distribution of Frequent Cold")

plt.tight_layout()
plt.show()
```
